```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#fungsi menghitung posisi proyekti pada waktu t
def posisi_proyektil(v0, theta, t, ho=0):
    #mengubah sudut dari derajat ke dalam radian
    theta = np.radians(theta)
    #percepatan gravitasi
    g = 9.81
    
    #menghitung kecepatan awal per kompoenen
    v0x = v0 * np.cos(theta)
    if ho == 0 :
        v0y = v0 * np.sin(theta)
    elif theta < 15 :
        v0y = 0
    else :
        v0y = v0 * np.sin(theta)
    
    #menghitung posisi proyektil pada waktu t
    x = v0x * t #horizontal
    y = v0y * t - 0.5 * g * t**2 + ho #vertikal
    
    return x, y
```

```python
#fungsi menampilkan lintasan 
def plot_lintasan(v0,theta,ho=0):
    #membuat array dari waktu 0 sampai waktu saat proyektil mencapai tanah
    if ho ==0 :
        t_max = 2 * v0 *np.sin(np.radians(theta)) / 9.81
    else:
        v_t = np.sqrt(2*9.81*ho)
        t_max = v_t/9.81
    t = np.linspace(0, t_max, num=100)
    
    # menghitung posisi proyektil pada waktu 
    x_posisi = []
    y_posisi= []
    
    for time in t:
        x, y = posisi_proyektil(v0, theta, time, ho)
        x_posisi.append(x)
        y_posisi.append(y)
    
    #membuat plot
    plt.figure(figsize=(10,6))
    plt.plot(x_posisi, y_posisi, 'b-', label='Lintasan Proyektil')
    plt.grid(True)
    plt.xlabel('Jarak horizontal(m)')
    plt.ylabel('Ketinggian(m)')
    plt.title('Lintasan Proyektil dengan Kecepatan Awal {} m/s dan Sudut {}'.format(v0, theta))
    plt.legend()
    plt.axis('equal')
    plt.show()
    
```

```python
v0 = 20
theta = 45
plot_lintasan(v0, theta)
```

```python
# vo tetap = 20, sudut = 15, 30, 45, 60, 75, 90
V0 = 20
theta = [15, 30, 45, 60, 75, 90]
for t in theta:
    plot_lintasan(v0, t)
```

```python
#theta tetap = 45, vo = 10, 20, 30, 40, 50
V0 = [10, 20, 30, 40, 50]
theta = 45
for v in V0:
    plot_lintasan(v, theta)
```

```python
#kombiinasi vo = 10, 20, 30, 40, 50, sudut = 15, 30, 45, 60, 75, 90
V0 = [10, 20, 30, 40, 50]
theta = [15, 30, 45, 60, 75, 90]
for v in V0:
    for t in theta:
        plot_lintasan(v, t)
```
