# Kasus Khusus pada limit

```python
# limit tak hingga 
def f_tak_hingga(x):
    return 1/(x-1)
```

```python
Lim_cal = LimitCalculator()
Lim_cal.plot_function(f_tak_hingga,1)
print(Lim_cal.analyze_limit_existence(f_tak_hingga,1))
print(f'limit kanan {Lim_cal.calculate_right_limit(f_tak_hingga,1)}')
print(f'limit kiri {Lim_cal.calculate_left_limit(f_tak_hingga,1)}')
```

```python
#fungsi osilasi
def f_osilasi(x):
    return np.sin(1/x)
```

```python
Lim_cal = LimitCalculator()
Lim_cal.plot_function(f_osilasi,0)
print(Lim_cal.analyze_limit_existence(f_osilasi,0))
print(f'limit kanan {Lim_cal.calculate_right_limit(f_osilasi,0)}')
print(f'limit kiri {Lim_cal.calculate_left_limit(f_osilasi,0)}')
```

```python
#limit jump diskontinuitas
def f_jump(x):
    if x < 0:
        return 1
    else:
        return 2
```

```python
Lim_cal = LimitCalculator()
Lim_cal.plot_function(f_jump,0)
print(Lim_cal.analyze_limit_existence(f_jump,0))
print(f'limit kanan {Lim_cal.calculate_right_limit(f_jump,0)}')
print(f'limit kiri {Lim_cal.calculate_left_limit(f_jump,0)}')
```

```python
#limit 0/0
def f_0_0(x):
    return (x**2 - 4)/(x-2)
```

```python
Lim_cal = LimitCalculator()
Lim_cal.plot_function(f_0_0,2)
print(Lim_cal.analyze_limit_existence(f_0_0,2))
print(f'limit kanan {Lim_cal.calculate_right_limit(f_0_0,2)}')
print(f'limit kiri {Lim_cal.calculate_left_limit(f_0_0,2)}')
```

```python
#limit tak hingga/tak hingga
def f_tak_hingga_2(x):
    return (x**2 - 1)/(x**2-1)
```

```python
Lim_cal = LimitCalculator()
Lim_cal.plot_function(f_tak_hingga_2,100000)
print(Lim_cal.analyze_limit_existence(f_tak_hingga_2,100000))
print(f'limit kanan {Lim_cal.calculate_right_limit(f_tak_hingga_2,100000)}')
print(f'limit kiri {Lim_cal.calculate_left_limit(f_tak_hingga_2,100000)}')
```
